```python
def afficher_menu():
  print("1. Ajouter une matrice")
  print("2. Afficher le maximum d'une matrice")
  print("3. Afficher le minimum d'une matrice")
  print("4. Afficher la moyenne d'une matrice")
  print("5. Afficher la somme d'une matrice")
  print("6. Quitter")

def ajouter_matrice():
  lignes = int(input("Entrez le nombre de lignes : "))
  colonnes = int(input("Entrez le nombre de colonnes : "))
  matrice = []
  for i in range(lignes):
    ligne = []
    for j in range(colonnes):
      element = int(input(f"Entrez l'élément [{i}][{j}] : "))
      ligne.append(element)
    matrice.append(ligne)
  matrices.append(matrice)
  print("Matrice ajoutée avec succès !")

def afficher_maximum(matrice):
  if not matrice:
    print("La matrice est vide.")
    return
  maximum = matrice[0][0]
  for ligne in matrice:
    for element in ligne:
      if element > maximum:
        maximum = element
  print("Le maximum de la matrice est :", maximum)

def afficher_minimum(matrice):
  if not matrice:
    print("La matrice est vide.")
    return
  minimum = matrice[0][0]
  for ligne in matrice:
    for element in ligne:
      if element < minimum:
        minimum = element
  print("Le minimum de la matrice est :", minimum)

def afficher_moyenne(matrice):
  if not matrice:
    print("La matrice est vide.")
    return
  somme = 0
  nombre_elements = 0
  for ligne in matrice:
    for element in ligne:
      somme += element
      nombre_elements += 1
  moyenne = somme / nombre_elements
  print("La moyenne de la matrice est :", moyenne)

def afficher_somme(matrice):
  if not matrice:
    print("La matrice est vide.")
    return
  somme = 0
  for ligne in matrice:
    for element in ligne:
      somme += element
  print("La somme de la matrice est :", somme)

matrices = []
while True:
  afficher_menu()
  choix = input("Entrez votre choix : ")
  if choix == '1':
    ajouter_matrice()
  elif choix == '2':
    if matrices:
      afficher_matrices()
      index = int(input("Entrez l'index de la matrice : "))
      if 0 <= index < len(matrices):
        afficher_maximum(matrices[index])
      else:
        print("Index de matrice non valide.")
    else:
      print("Aucune matrice à afficher.")
  elif choix == '3':
    if matrices:
      afficher_matrices()
      index = int(input("Entrez l'index de la matrice : "))
      if 0 <= index < len(matrices):
        afficher_minimum(matrices[index])
      else:
        print("Index de matrice non valide.")
    else:
      print("Aucune matrice à afficher.")
  elif choix == '4':
    if matrices:
      afficher_matrices()
      index = int(input("Entrez l'index de la matrice : "))
      if 0 <= index < len(matrices):
        afficher_moyenne(matrices[index])
      else:
        print("Index de matrice non valide.")
    else:
      print("Aucune matrice à afficher.")
  elif choix == '5':
    if matrices:
      afficher_matrices()
      index = int(input("Entrez l'index de la matrice : "))
      if 0 <= index < len(matrices):
        afficher_somme(matrices[index])
      else:
        print("Index de matrice non valide.")
    else:
      print("Aucune matrice à afficher.")
  elif choix == '6':
    break
  else:
    print("Choix non valide.")
```
